```python
import networkx as nx
import numpy as np
import itertools as it
import random
import time
import scipy
import fresnel
import os
import PIL
import glob

from matplotlib import pyplot as plt
k_on_Ag = 1.0
```

```python
step = 500000
estimated_dipole_num = []
atom_number = []
for k_Au in np.around(np.linspace(-4,-6,11),5):
    for k_Ag in np.around(np.linspace(-4,-6,11),5):
        for Energy_ratio in np.around(np.linspace(0.1,1,4),5):
            for Energy_ratio2 in np.around(np.linspace(0.1,1,4),5):
                for k_on_Au in np.around(np.linspace(0.1,1,4),5):
                    path="/mnt/STORAGE2//Nanobot-Monte_Carlo_Simulation/Au_Ag_7//example_{}_{}_{}_{}_{}_{}/".format(k_Au,
                                                                                                     k_Ag,
                                                                                                     Energy_ratio,
                                                                                                     Energy_ratio2,
                                                                                                     k_on_Au,
                                                                                                     k_on_Ag)
                    print(path)
                    try:
                        Au_data=np.loadtxt(path+"/structure/"+f"/Au_data_{step}.csv",delimiter=",").reshape(-1,5)
                        Ag_data=np.loadtxt(path+"/structure/"+f"/Ag_data_{step}.csv",delimiter=",").reshape(-1,5)

                        estimated_dipole_num.append(len(np.unique(np.around(np.vstack((Au_data[:,0:3], Ag_data[:,0:3]))),axis=0)))
                        atom_number.append([len(Au_data),len(Ag_data)])
                    except:
                        pass
```
