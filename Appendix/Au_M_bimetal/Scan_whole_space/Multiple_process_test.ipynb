```python
import os
import subprocess
import multiprocessing
import numpy as np
import time
```

```python
def run_single_simulation(k_Au=-2,
                          k_Ag=-2,
                          Energy_ratio=0.5,
                          Energy_ratio2=0.5,
                          k_on_Au=0.1,
                          k_on_Ag=1.0):
    
    f = open("Ag_Au_template2.py", "r")
    list_of_lines=f.readlines()
    #value for:
    #k_Au,k_Ag,Energy_ratio,k_on_Ag,k_on_Au
    data=[k_Au,k_Ag,Energy_ratio,Energy_ratio2,k_on_Au,k_on_Ag]
    for i in range(22,28):
        print(list_of_lines[i])
        list_of_line_element=list_of_lines[i].split("=")
        list_of_line_element[-1]="%f"%data[i-22]
        list_of_lines[i]=list_of_line_element[0]+"="+list_of_line_element[-1]+"\n"
        print(list_of_lines[i])

    path="/mnt/orkney/Chemobot/Nanobot-Monte_Carlo_Simulation/Au_Ag_7/example_{}_{}_{}_{}_{}_{}/".format(k_Au,
                                                                                                         k_Ag,
                                                                                                         Energy_ratio,
                                                                                                         Energy_ratio2,
                                                                                                         k_on_Au,
                                                                                                         k_on_Ag)
    if os.path.isdir(path)==False:
        os.mkdir(path)
        os.mkdir(path+"structure")
        os.mkdir(path+"plot")
        
    path_name=path+"example_{}_{}_{}_{}_{}_{}.py".format(*data)
    file1 = open(path_name,"w") 
    file1.writelines(list_of_lines) 
    file1.close()

    subprocess.run("python3 "+path_name, shell=True)
```

```python
if __name__ == '__main__':
    pool = multiprocessing.Pool(40) #use all available cores, otherwise specify the number you want as an argument
    for k_Au in np.around(np.linspace(-2,-4,11),5)[5:]:
        for k_Ag in np.around(np.linspace(-4,-6,11),5):
            for Energy_ratio in np.around(np.linspace(0.1,1,4),5):
                for Energy_ratio2 in np.around(np.linspace(0.1,1,4),5):
                    for k_on_Au in np.around(np.linspace(0.1,1,4),5):
                        pool.apply_async(run_single_simulation, args=(k_Au,k_Ag,Energy_ratio,Energy_ratio2,k_on_Au,1.0))
                        time.sleep(1)
    pool.close()
    pool.join()
```
