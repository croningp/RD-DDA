```python
import matplotlib.image as mpimg
import numpy as np 
from matplotlib import pyplot as plt
import fresnel
import PIL
k_on_Ag = 1.0
```

```python
cpu_limit = fresnel.Device(mode='cpu', n=10)
for k_Au in np.around(np.linspace(-4,-6,11),5):
    for k_Ag in np.around(np.linspace(-4,-6,11),5):
        for Energy_ratio in np.around(np.linspace(0.1,1,4),5):
            for Energy_ratio2 in np.around(np.linspace(0.1,1,4),5):
                for k_on_Au in np.around(np.linspace(0.1,1,4),5):
                    path="/mnt/STORAGE2//Nanobot-Monte_Carlo_Simulation/Au_Ag_7//example_{}_{}_{}_{}_{}_{}/".format(k_Au,
                                                                                                     k_Ag,
                                                                                                     Energy_ratio,
                                                                                                     Energy_ratio2,
                                                                                                     k_on_Au,
                                                                                                     k_on_Ag)
                    step = 500000
                    try:
                        step = 500000
                        Au = np.loadtxt(path+f'/structure/Au_data_{step}.csv',delimiter=',')
                        Ag = np.loadtxt(path+f'/structure/Ag_data_{step}.csv',delimiter=',')
                        scene = fresnel.Scene(device=cpu_limit)
                        geometry1 = fresnel.geometry.Sphere(scene, N=len(Au)+len(Ag), radius=np.sqrt(2)/4)
                        geometry1.material = fresnel.material.Material(color=fresnel.color.linear([1,1,0]),
                                                                      roughness=0.8)
                        geometry1.position[:len(Au)] = Au[:,0:3]
                        geometry1.material.primitive_color_mix = 1

                        geometry1.color[0:len(Au)]=fresnel.color.linear(np.array([1,1,0]))
                        if len(Ag)>0:
                            Ag = Ag.reshape(-1,5)
                            geometry1.color[len(Au):]=fresnel.color.linear(np.array([0,0,1]))
                            geometry1.position[len(Au):] = Ag[:,0:3]

                        scene.camera = fresnel.camera.Orthographic(position=(60,-60,30), look_at=(0,0,10),up=(0,0,1), height=60)

                        fresnel.preview(scene)
                        out = fresnel.preview(scene)
                        image = PIL.Image.fromarray(out[:], mode='RGBA')
                        image.save(f"/mnt/scapa/group/Yibin Jiang/Au_Ag_7/{step}/example_{k_Au}_{k_Ag}_{Energy_ratio}_{Energy_ratio2}_{k_on_Au}_{k_on_Ag}_{step}.png",
                                  dpi=(500, 500))

                    except:
                        print(f'{path} is not found')
```
